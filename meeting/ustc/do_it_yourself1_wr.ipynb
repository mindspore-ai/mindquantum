```python
n = 10
c = Circuit()
c += UN(H, n)
for i in range(n):
    c += RZ({'theta': 2**i}).on(n, n-i-1)
print(c)
```

```python
c += qft(range(n)).hermitian
print(c)
```

```python
sim = Simulator('projectq', c.n_qubits)
```

```python
sim.apply_circuit(c, {'theta': -4*np.pi*0.1234})
qs=sim.get_qs()
qs
```

```python
index = np.argmax(np.abs(qs))
print(index)
bit_string = bin(index)[2:].zfill(c.n_qubits)[1:]
print(bit_string)
```

```python
theta_exp = int(bit_string[::-1], 2) / 2**n
theta_exp
```
