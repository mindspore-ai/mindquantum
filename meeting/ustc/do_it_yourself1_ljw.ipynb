# Do it yourself
验证均匀叠加态经过傅里叶变换后的量子态为零态

```python
my_circuit = Circuit([H.on(0), H.on(1), H.on(2), H.on(3)])
my_circuit += qft(range(4))
print(my_circuit)
```

```python
my_sim  = Simulator('projectq', 4)
my_sim.apply_circuit(my_circuit)
my_sim.get_qs()
```

```python
n = 10
c = Circuit()
c += UN(H, n)
for i in range (n):
    c += RZ({'theta': 2**i}).on(n, n-i-1)
print(c)
```

```python
c += qft(range(n)).hermitian
print(c)
```

```python
sim = Simulator('projectq', c.n_qubits)
```

```python
sim.apply_circuit(c, {'theta': -4*np.pi*0.1234})
```
